```
# difference between sort by and order by
# 1.Order By and Sort By both are not same in sql. Order by will do sorting an entire data. sort by will do partition wise sorting
# 2.orderBy and sort both are same in pyspark. sortWintinPartitions as same as sort By in sql.
#In PySpark:
# order by=sort 
#sort by=sortWintinPartitions
#In PySQL:
#Order by entire data set level
#Sort by partition level
```

```
from pyspark.sql.functions import spark_partition_id
df = spark.read.format("delta").load("/user/hive/warehouse/delta_join_table").repartition(4,"DEPTNO").withColumn("PARTITION_ID",spark_partition_id())
df.createOrReplaceTempView("df")
display(df)
```

```
%sql
select * from df order by sal
```

```
%sql
select * from df sort by sal
```

```
df.orderBy("SAL").show()
```

```
df.sort("SAL").show()
```

```
df.sortWithinPartitions("SAL").show()
```
